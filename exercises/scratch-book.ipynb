Multiple Sequence Alignment

### 

| <div style="width:200px"></div>        | <div style="width:100px"></div>   |
|--------------------------------------------|-----------------|
|  Binary Search                      | $O(\log_2 N)$   |
|  Quick Sort                         | $O(N\log_2 N)$  |
|  Fast Fourier Transform (FFT)  | $O(N\log_2 N)$  |

```python
import math

def invsqr_sum(n):
    return sum(i**(-2) for i in range(1, n+1))

def invsqr_sum_approx(n):
    res = 1/(n+0.5) - 1.0/12/(n+0.5)**3 + 0.0/35.5/(n+0.5)**5
    return math.pi**2/6 - res

def invsqr_sum_approx2(n):
    ''' Eurler-Maclaurin '''
    res = 1/(n+1) + 0.5/(n+1)**2 + 1.0/6/(n+1)**3 -1/30/(n+1)**5
    return math.pi**2/6 - res

x, y, y2 = [], [], []
for n in range(6, 100):
    err = invsqr_sum_approx(n) - invsqr_sum(n)
    err2 = invsqr_sum_approx2(n) - invsqr_sum(n)
    #print(n, err*(n+0.5)**5, err2*(n+0.5)**5)
    x.append(n)
    y.append(abs(err))
    y2.append(abs(err2))

print(max(y), max(y2))

import matplotlib.pyplot as plt

plt.plot(x, y, color='b')
plt.plot(x, y2, color='r')
plt.show()
```
